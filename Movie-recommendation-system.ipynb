# Data preprocessing 

```python
# checking missing values 
movies.isna().sum()
```

```python
movies = movies.dropna()
```

```python
movies.isnull().sum()
```

```python
# lets find any duplicate values exists in the data 
movies.duplicated().sum()
```

```python
import ast 
```

```python
movies.iloc[0].genres
```

```python
ast.literal_eval(movies.iloc[0].genres)
```

```python
# create a helper fxn to create tags and do data preprocessing 
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L        
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres'].head()
```

```python
movies.head()
```

```python
# clean the keywords column 
movies['keywords'][0]
```

```python
# apply convert function here also 
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies['keywords'][0]
```

```python
movies.head()
```

```python
# convert the cast column into a list of tags 
movies['cast'].iloc[0]
```

```python
# keep main top 3 actors name from the cast column 
def convert_3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else: 
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_3)
```

```python
movies.iloc[0]['cast']
```

```python
movies.head()
```

```python
# fetch director name from crew column 
movies['crew'][0]
```

```python
# helper fxn to fetch the directors name 
def get_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
movies['overview'].iloc[0]
```

```python
# overview is a string -> convert it into list 
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head()
```

```python
# sam worthington and sam mendes both are different but sam is kept same for both and recommended sam worthington with sam mended
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(' ', '') for i in x] )
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(' ', '') for i in x] )
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(' ', '') for i in x] )
```

```python
movies.head() # Science Fiction -> ScienceFiction
```

```python
# combine all the columns into one [tags]
movies['tags'] = movies['overview'] + movies['keywords']+ movies['genres'] + movies['cast'] + movies['crew']
movies['tags'][0]
```

```python
movies.head()
```

```python
# create a new df with tags , movie id , title 
df = movies[['movie_id','title','tags']]
df.head()
```

```python
# convert tags list into strings dtype
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df['tags'][0]
```

```python
# recommended to convert tags into lower case 
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
df.head()
```

# VECTORIZATION

```python
# using vectorization to convert tags into vector of words frequency 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')
vect = cv.fit_transform(df['tags'])
```

```python
# create object 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# create helper function 
def stem(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
df['tags'] = df['tags'].apply(stem)
```

# MAIN FUNCTION

```python
# use cosine distance to measure similarity between vectors for recommendations 
from sklearn.metrics.pairwise import cosine_similarity # distance incs , smilarity decs (sim = 0 to1)
similarity = cosine_similarity(vect)
```

```python
similarity.shape
```

step 1 : movies index find <br>
step 2 : using movies index go to similarity matrix  to dinf distances <br>
step 3 : sort the distances with movies index and take first 9 similar movies <br>
step 4 : print the names of the movies<br>

```python
def recommend(movie):
    movie_index = df[df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key = lambda x: x[1])[1:10]

    for i in movies_list:
        print(df['title'].iloc[i[0]])
```
